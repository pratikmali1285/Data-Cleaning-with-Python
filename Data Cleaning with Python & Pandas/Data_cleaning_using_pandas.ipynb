```python
import pandas as pd 
```

```python
df = pd.read_csv("students_data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
df.loc[:,"name"] = df["name"].str.lower().str.title()
df
```

```python
df["age"].unique()
```

```python
age_mean = df["age"].mean().round(0)
```

```python
df.loc[:,"age"] = df["age"].fillna(age_mean)
```

```python
df
```

```python
df["gender"].unique()
```

```python
df.loc[:,"gender"] = df["gender"].str.lower().str.title()
df
```

```python
df["grade"].unique()
```

```python
df.loc[:, "grade"] = df["grade"].replace({'10th':'10','11th':'11'})
```

```python
df
```

```python
math_mean = df['math_score'].mean()
```

```python
df.loc[:,"math_score"] = df["math_score"].fillna(math_mean)
```

```python
df
```

```python
df["english_score"]
```

```python
df.loc[:, "english_score"] = pd.to_numeric(df["english_score"], errors='coerce')
```

```python
df
```

```python
eng_mean = df['english_score'].mean()
eng_mean1 = round(eng_mean, 0)  # or round(eng_mean)

df.loc[:,"english_score"] = df["english_score"].fillna(eng_mean1)
```

```python
df.loc[:,"english_score"] = df["english_score"].round(0)
```

```python
df
```

```python
df["science_score"]
```

```python
df.loc[:,"science_score"] = df["science_score"].dropna()
```

```python
df
```

```python
df["enrolled_date"].unique()
```
